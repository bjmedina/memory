```python
pairs_path = "/om2/user/bjmedina/data/texture_pairs/max_cochleagram_mse-time_avg_mse-25.0db-lowpassfiltered-10.0hz-strict-puretone-gt40-check-lessthan-0.62-offset0.1-segment_duration0.5/all_pairs/"
all_sounds = glob.glob(pairs_path + "*")
sounds_to_save = all_sounds[:total_pairs_of_repeat_stimuli*4]

to_save_path = "/mindhive/mcdermott/www/mturk_stimuli/bjmedina/texture_pairs/discriminable_pairs"


to_save = []

for i, sound_to_save in enumerate(sounds_to_save):
    new_fname = f"mem_stim_num{i//2 + 1}_{i%2+1}.wav"
    full_path = os.path.join(to_save_path, new_fname)

    print(sound_to_save.split("/")[-1], "==>", new_fname)

    # Save metadata
    to_save_dict = {
        'filename': os.path.basename(sound_to_save),
        'full_path': sound_to_save,
        'clip_source_full_path': full_path,
        'clip_source': new_fname
    }
    to_save.append(to_save_dict)

    # Actually copy the file
    shutil.copyfile(sound_to_save, full_path)

json_path = os.path.join(to_save_path, "filenames.json")

with open(json_path, 'w') as f:
    json.dump(to_save, f, indent=2)
```

```python
def read_json_file(file_path):
    """
    Opens and reads a JSON file, then returns the data as a Python object (usually a dictionary or list).
    """
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            return data
    except FileNotFoundError:
        print(f"Error: File not found at path: {file_path}")
        return None
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON format in file: {file_path}")
        return None

# Example usage:
file_path = '/mindhive/mcdermott/www/mturk_stimuli/bjmedina/texture_pairs/discriminable_pairs/filenames.json'  # Replace 'data.json' with the actual path to your JSON file
data = read_json_file(file_path)
print(*data, sep='\n')
```
