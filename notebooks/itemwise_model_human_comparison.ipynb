# Item-wise model vs human comparison

Runs the model in itemwise mode and correlates model FA rates with human FA rates per item.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
from utls.runners_v2 import run_experiment_scores_prior

```

```python
# Run item-wise scores
out = run_experiment_scores_prior(
    sigma0=sigma0_fitted,
    noise_mode=noise_mode,
    metric=metric,
    X0=X,
    name_to_idx=name_to_idx,
    experiment_list=experiment_list,
    score_model=score_fn,
    drift_step_size=drift_step_size_fitted,
    return_item_scores=True,
)

```

```python
# Convert model item FAs into rates
model_fa = {k: float(np.mean(v)) if len(v) else np.nan for k, v in out['item_fas'].items()}

```

```python
# Join with human data table (provide `human_item_fa` dict)
common = sorted(set(model_fa) & set(human_item_fa))
x = np.array([model_fa[k] for k in common], dtype=float)
y = np.array([human_item_fa[k] for k in common], dtype=float)
mask = np.isfinite(x) & np.isfinite(y)
rho, p = spearmanr(x[mask], y[mask])
plt.figure(figsize=(5,5))
plt.scatter(x[mask], y[mask], alpha=0.7)
plt.xlabel('Model FA rate')
plt.ylabel('Human FA rate')
plt.title(f'Item-wise FA: Spearman r={rho:.3f}, p={p:.3g}')
plt.show()

```
