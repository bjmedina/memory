```python
import torch
import numpy as np
from scipy.io import savemat
import sys
sys.path.append('/om2/user/jmhicks/projects/TextureStreaming/code/')
from texture_prior.utils import prior
from texture_prior.utils import path
from texture_prior.utils import projection
import matplotlib.pyplot as plt
```

```python
filename = path.relative('../assets/principal_components.pt')
print(filename)
PC_info = torch.load(filename)
varex = torch.cumsum(PC_info['variances'], 0) / PC_info['variances'].sum()
plt.figure()
plt.plot(torch.cumsum(PC_info['variances'], 0) / PC_info['variances'].sum())
plt.xlim(0,2000)
plt.show()

nPCs = [256, 512, 1024, 2048, 4096]
for n in nPCs:
    print(varex[n-1])
```

```python
stat_projector = projection.statistics_projector() # setup stat projector
stats_file = path.relative('../assets/texture_statistics.pt')
stats_s = torch.load(stats_file)
data_s = stat_projector.project(stats_s, normalize=True, nPCs=4096) # transform into PC space
print(stats_s.shape)
print(data_s.shape)
save_name = '/om2/user/lakshmin/audio-prior/assets/texture_statistics_4096PCs.pt'
torch.save(data_s, save_name)
```
