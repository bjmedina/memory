```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decay_curves(noise_std=1.0, linear_rate=0.01, exp_rate=0.01, power_beta=0.5, max_age=200):
    """
    Compare linear, exponential, and power-law growths of noise std (σ_t)
    and their corresponding variances (σ_t^2).

    Args:
        noise_std (float): base σ0
        linear_rate (float): α for linear/sqrt growth
        exp_rate (float): α for exponential growth exponent
        power_beta (float): β exponent for power-law growth
        max_age (int): maximum trace age to plot
    """
    ages = np.arange(max_age + 1, dtype=float)
    ages[0] = 1e-6  # avoid zero in exponentiation

    # --- noise std schedules ---
    sigma_linear = noise_std * np.sqrt(ages)   # √t growth
    sigma_exp    = noise_std * np.exp(ages ** exp_rate)           # e^{αt}
    sigma_power  = noise_std * (ages ** power_beta)              # t^β

    # --- variances ---
    var_linear = sigma_linear**2
    var_exp    = sigma_exp**2
    var_power  = sigma_power**2

    # --- Plot both std and variance ---
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    # σ_t
    axes[0].plot(ages, sigma_linear, label=f"√t (β=0.5)", lw=2)
    axes[0].plot(ages, sigma_exp, label=f"Exponential (α={exp_rate})", lw=2)
    axes[0].plot(ages, sigma_power, label=f"Power-law (β={power_beta})", lw=2)
    axes[0].set_xlabel("Trace age (timesteps)")
    axes[0].set_ylabel("Noise std σ(t)")
    axes[0].set_title("Noise standard deviation over time")
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    axes[0].axvline(x=16, linestyle='--', color='k', alpha=0.5)
    axes[0].axvline(x=32, linestyle='--', color='k', alpha=0.5)
    axes[0].axvline(x=64, linestyle='--', color='k', alpha=0.5)
    axes[0].axvline(x=128, linestyle='--', color='k', alpha=0.5)

    # σ_t^2
    axes[1].plot(ages, var_linear, label=f"√t → var ∝ t", lw=2)
    axes[1].plot(ages, var_exp, label=f"Exponential → var ∝ e^(2αt)", lw=2)
    axes[1].plot(ages, var_power, label=f"Power-law → var ∝ t^{2*power_beta}", lw=2)
    axes[1].set_xlabel("Trace age (timesteps)")
    axes[1].set_ylabel("Noise variance σ(t)²")
    axes[1].set_title("Noise variance growth over time")
    axes[1].legend()
    axes[1].grid(alpha=0.3)
    axes[1].axvline(x=16, linestyle='--', color='k', alpha=0.5)
    axes[1].axvline(x=32, linestyle='--', color='k', alpha=0.5)
    axes[1].axvline(x=64, linestyle='--', color='k', alpha=0.5)
    axes[1].axvline(x=128, linestyle='--', color='k', alpha=0.5)

    plt.suptitle("Comparison of Noise Growth Schedules (σ_t and σ_t²)", y=1.03)
    plt.tight_layout()
    plt.show()

# Example
plot_decay_curves(noise_std=0.2, linear_rate=0.01, exp_rate=0.15, power_beta=0.45, max_age=128)
```

```python
import numpy as np
import matplotlib.pyplot as plt

ages = np.linspace(1, 32, 50)
noise_std = 1.0
exp_rate   = 0.25
power_beta = 0.45
t0 = 1  # reference time (ISI=0 equivalent)
sigma_target = 1.0

# raw schedules
sigma_linear = noise_std * np.sqrt(ages)
sigma_exp    = noise_std * np.exp(ages ** exp_rate)
sigma_power  = noise_std * (ages ** power_beta)

# normalize each schedule so σ(t0) = sigma_target
sigma_linear *= sigma_target / (noise_std * np.sqrt(t0))
sigma_exp    *= sigma_target / (noise_std * np.exp(t0 ** exp_rate))
sigma_power  *= sigma_target / (noise_std * (t0 ** power_beta))

plt.figure(figsize=(6,4))
plt.plot(ages, sigma_linear, label="√t")
plt.plot(ages, sigma_exp, label=f"exp (α={exp_rate})")
plt.plot(ages, sigma_power, label=f"power (β={power_beta})")
plt.axvline(t0, color='k', linestyle='--', alpha=0.5)
plt.legend()
plt.xlabel("Age (timesteps)")
plt.ylabel("σ(t)")
plt.title("Normalized noise schedules (σ_t0 = constant)")
plt.tight_layout()
plt.show()
```
