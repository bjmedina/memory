```python
from IPython.display import display, clear_output
import IPython.display as ipd
import random

def batched_selection(stims, keep_list, label="nhs"):
    while len(keep_list) < 46 and stims:
        clear_output(wait=True)
        print(f"Category: {label}")
        print(f"Selected: {len(keep_list)} / 46")
        print("Showing 20 random sounds...\n")

        # Step 1: Choose a batch of up to 20
        batch = random.sample(stims, min(20, len(stims)))

        # Step 2: Display the batch
        for stim in batch:
            print(stim['stim_path'])
            display(ipd.Audio(stim['stim_path']))

        # Step 3: Get user input
        choice = input("\nType a filename to KEEP, 'next' for new batch, or 'exit' to stop: ").strip()

        if choice.lower() == 'exit':
            print("Exiting selection.")
            break
        elif choice.lower() == 'next':
            continue

        # Step 4: Match filename and store if valid
        matched = [s for s in batch if s['stim_path'] == choice]
        if matched:
            keep_list.append(matched[0])
            stims.remove(matched[0])
        else:
            print(f"Filename '{choice}' not found in this batch. Try again.")

    clear_output(wait=True)
    print(f"Done selecting from {label}. Final count: {len(keep_list)}")
```

```python
import shutil
def prepare_stimuli(list_a, list_b, out_dir, base_name="mem_stim", json_name="filenames.json"):
    """
    Combines and renames stimuli from two lists, alternating entries (even from list_a, odd from list_b),
    copies files into a new directory, and generates a JSON index file.

    Args:
        list_a (list): First list of dictionaries (assigned even indices)
        list_b (list): Second list of dictionaries (assigned odd indices)
        out_dir (str): Directory to save new stimuli and the JSON
        base_name (str): Base filename prefix (default: 'mem_stim')
        json_name (str): Name of the JSON file to save (default: 'filenames.json')
    """
    os.makedirs(out_dir, exist_ok=True)
    combined = []
    idx = 0
    a_idx, b_idx = 0, 0

    while a_idx < len(list_a) or b_idx < len(list_b):
        if idx % 2 == 0 and a_idx < len(list_a):
            item = list_a[a_idx]
            a_idx += 1
        elif idx % 2 == 1 and b_idx < len(list_b):
            item = list_b[b_idx]
            b_idx += 1
        else:
            idx += 1
            continue

        new_filename = f"{base_name}_{idx}.wav"
        new_stim_path = os.path.join(out_dir, new_filename)

        # Copy the file
        shutil.copy(item['stim_path'], new_stim_path)

        # Construct new entry
        new_entry = {
            'orig_path': item['stim_path'],
            'stim_path': new_stim_path,
            'filename': new_filename,
            'onset': item.get('onset', 0.0),
            'type': item['type']
        }

        combined.append(new_entry)
        idx += 1

    # Save JSON
    json_path = os.path.join(out_dir, json_name)
    with open(json_path, 'w') as f:
        json.dump(combined, f, indent=2)

    print(f"✅ Saved {len(combined)} files to {out_dir}")
    print(f"📝 JSON saved to: {json_path}")
```
