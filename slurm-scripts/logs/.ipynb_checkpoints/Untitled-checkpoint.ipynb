```python
# Notebook analysis for multi-start optimization logs (Powell/ZScore only)
# Loads latest logs from a directory or explicit files, summarizes, and plots.
# Minimal deps: numpy, pandas, matplotlib. Optional: ipywidgets for interactivity.

# ---------------------------
# Imports (top; required)
# ---------------------------
import os, glob, json, re
from typing import Optional, Tuple, Dict
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.tri import Triangulation

# Optional widgets
try:
    import ipywidgets as W
    from IPython.display import display
    _HAS_WIDGETS = True
except Exception:
    _HAS_WIDGETS = False


# ---------------------------
# IO + selection
# ---------------------------
def latest_with_prefix(d: str, prefix: str) -> Optional[str]:
    """Newest file in dir matching '<prefix>_*.csv'."""
    pats = sorted(glob.glob(os.path.join(d, f"{prefix}_*.csv")))
    return pats[-1] if pats else None

def load_logs(
    directory: Optional[str] = None,
    eval_csv: Optional[str] = None,
    runs_csv: Optional[str] = None,
    marginal_nv_csv: Optional[str] = None,
    marginal_cr_csv: Optional[str] = None,
) -> Tuple[pd.DataFrame, pd.DataFrame, Optional[pd.DataFrame], Optional[pd.DataFrame], str]:
    """Load eval/runs/marginal CSVs; infer latest from directory if paths are None."""
    if directory:
        eval_csv = eval_csv or latest_with_prefix(directory, "eval_log")
        runs_csv = runs_csv or latest_with_prefix(directory, "runs_summary")
        marginal_nv_csv = marginal_nv_csv or latest_with_prefix(directory, "marginal_noise_variance")
        marginal_cr_csv = marginal_cr_csv or latest_with_prefix(directory, "marginal_criterion")
    if not eval_csv or not runs_csv:
        raise FileNotFoundError("Need eval_csv and runs_csv (or provide directory with those files).")
    eval_df = pd.read_csv(eval_csv)
    runs_df = pd.read_csv(runs_csv)
    nv_df = pd.read_csv(marginal_nv_csv) if marginal_nv_csv and os.path.exists(marginal_nv_csv) else None
    cr_df = pd.read_csv(marginal_cr_csv) if marginal_cr_csv and os.path.exists(marginal_cr_csv) else None
    save_dir = directory or os.path.dirname(os.path.abspath(eval_csv)) or "."
    return eval_df, runs_df, nv_df, cr_df, save_dir


# ---------------------------
# Checks + summary
# ---------------------------
def coerce_eval_schema(df: pd.DataFrame) -> pd.DataFrame:
    """Ensure columns: noise_variance, criterion, mse (rename from 'nv' if needed)."""
    if "nv" in df.columns and "noise_variance" not in df.columns:
        df = df.rename(columns={"nv": "noise_variance"})
    need = {"noise_variance", "criterion", "mse"}
    missing = need - set(df.columns)
    if missing:
        raise ValueError(f"Missing columns in eval_df: {missing}")
    return df

def compute_summary(eval_df: pd.DataFrame, runs_df: pd.DataFrame) -> Dict:
    """Basic stats and global best row from eval_df."""
    best_idx = int(eval_df["mse"].argmin())
    best_row = eval_df.iloc[best_idx].to_dict()
    stats = {
        "n_evals": int(len(eval_df)),
        "n_runs": int(len(runs_df)),
        "mse_min": float(eval_df["mse"].min()),
        "mse_median": float(eval_df["mse"].median()),
        "mse_mean": float(eval_df["mse"].mean()),
    }
    return {"stats": stats, "best_eval_row": best_row}


# ---------------------------
# Marginalization
# ---------------------------
def marginalize_mse(eval_df: pd.DataFrame, param_col: str, n_bins: int = 30) -> pd.DataFrame:
    """Bin one param and aggregate MSE across the other (mean/median/min/count)."""
    assert param_col in ("noise_variance", "criterion")
    x = eval_df[param_col].to_numpy()
    y = eval_df["mse"].to_numpy()
    lo, hi = float(np.min(x)), float(np.max(x))
    if lo == hi:
        lo, hi = lo - 1e-6, hi + 1e-6
    edges = np.linspace(lo, hi, n_bins + 1)
    idx = np.clip(np.digitize(x, edges) - 1, 0, n_bins - 1)
    rows = []
    for b in range(n_bins):
        m = idx == b
        if not np.any(m): 
            continue
        vals = y[m]
        rows.append({
            f"{param_col}_bin_left": float(edges[b]),
            f"{param_col}_bin_right": float(edges[b+1]),
            f"{param_col}_bin_center": float(0.5 * (edges[b] + edges[b+1])),
            "mse_mean": float(np.mean(vals)),
            "mse_median": float(np.median(vals)),
            "mse_min": float(np.min(vals)),
            "count": int(vals.size),
        })
    return pd.DataFrame(rows).sort_values(f"{param_col}_bin_center").reset_index(drop=True)


# ---------------------------
# Plots (matplotlib only)
# ---------------------------
def plot_scatter(eval_df: pd.DataFrame):
    """All evaluations colored by MSE."""
    x, y, c = eval_df["noise_variance"], eval_df["criterion"], eval_df["mse"]
    plt.figure(figsize=(7.5, 6))
    sc = plt.scatter(x, y, c=c, s=14, cmap="viridis", edgecolors="none")
    plt.colorbar(sc, label="MSE")
    plt.xlabel("Noise variance")
    plt.ylabel("Criterion")
    plt.title("All evaluations")
    plt.tight_layout()
    plt.show()

def plot_contour(eval_df: pd.DataFrame):
    """Triangulated contour over irregular samples."""
    x = eval_df["noise_variance"].to_numpy()
    y = eval_df["criterion"].to_numpy()
    z = eval_df["mse"].to_numpy()
    tiny = 1e-9
    tri = Triangulation(x + tiny*np.random.randn(*x.shape), y + tiny*np.random.randn(*y.shape))
    plt.figure(figsize=(7.5, 6))
    tcf = plt.tricontourf(tri, z, levels=20, cmap="viridis")
    plt.colorbar(tcf, label="MSE")
    plt.xlabel("Noise variance")
    plt.ylabel("Criterion")
    plt.title("MSE contour")
    plt.tight_layout()
    plt.show()

def plot_trajectories(eval_df: pd.DataFrame, runs_df: pd.DataFrame):
    """Per-run optimizer paths; needs columns 'run' and 'call' in eval_df."""
    if not {"run", "call"} <= set(eval_df.columns):
        plot_scatter(eval_df); return
    plt.figure(figsize=(7.5, 6))
    plt.scatter(eval_df["noise_variance"], eval_df["criterion"], c="0.88", s=6, edgecolors="none", label="all evals")
    for _, row in runs_df.iterrows():
        rid = int(row["run"])
        sub = eval_df[eval_df["run"] == rid].sort_values("call")
        if len(sub) > 1:
            plt.plot(sub["noise_variance"], sub["criterion"], lw=0.8, alpha=0.8)
        if {"best_noise_variance","best_criterion"} <= set(runs_df.columns):
            plt.scatter(row["best_noise_variance"], row["best_criterion"], marker="*", s=90, edgecolors="k", facecolors="none")
    plt.xlabel("Noise variance"); plt.ylabel("Criterion"); plt.title("Optimization trajectories")
    plt.tight_layout(); plt.show()

def plot_param_hists(eval_df: pd.DataFrame):
    """Histograms of explored parameters."""
    fig, ax = plt.subplots(1, 2, figsize=(10, 4.2))
    ax[0].hist(eval_df["noise_variance"], bins=40)
    ax[0].set_xlabel("Noise variance"); ax[0].set_title("Explored noise variance")
    ax[1].hist(eval_df["criterion"], bins=40)
    ax[1].set_xlabel("Criterion"); ax[1].set_title("Explored criterion")
    fig.tight_layout(); plt.show()

def plot_marginal(df: pd.DataFrame, which: str, y_key: str = "mse_min"):
    """1D marginal line plot (pick mse_min or mse_mean)."""
    xcol = f"{which}_bin_center"
    if xcol not in df.columns and {f"{which}_bin_left", f"{which}_bin_right"} <= set(df.columns):
        df = df.copy()
        df[xcol] = 0.5 * (df[f"{which}_bin_left"] + df[f"{which}_bin_right"])
    y_key = y_key if y_key in df.columns else ("mse_mean" if "mse_mean" in df.columns else df.filter(like="mse_").columns[0])
    plt.figure(figsize=(7.5, 4.5))
    plt.plot(df[xcol], df[y_key], marker="o")
    plt.xlabel(which.replace("_", " ").title())
    plt.ylabel(y_key.replace("_", " ").upper())
    plt.title(f"Marginal {y_key} vs {which}")
    plt.tight_layout(); plt.show()


# ---------------------------
# Interactive widget (optional)
# ---------------------------
def interactive_run_view(eval_df: pd.DataFrame):
    """Widget to explore trajectories one run at a time."""
    if not _HAS_WIDGETS or not {"run","call"} <= set(eval_df.columns):
        print("ipywidgets not available or eval_df lacks 'run'/'call'; skipping.")
        return
    runs = sorted(eval_df["run"].unique().tolist())
    dd = W.Dropdown(options=runs, description="Run:", value=runs[0])
    out = W.Output()

    def render(run_id: int):
        sub = eval_df[eval_df["run"] == run_id].sort_values("call")
        with out:
            out.clear_output(wait=True)
            fig, ax = plt.subplots(1, 2, figsize=(12, 4.8))
            sc = ax[0].scatter(sub["noise_variance"], sub["criterion"], c=sub["mse"], s=28, cmap="viridis", edgecolors="none")
            fig.colorbar(sc, ax=ax[0], label="MSE")
            ax[0].set_xlabel("Noise variance"); ax[0].set_ylabel("Criterion"); ax[0].set_title(f"Run {run_id} evaluations")
            ax[1].plot(sub["call"], sub["mse"], marker="o")
            ax[1].set_xlabel("Call"); ax[1].set_ylabel("MSE"); ax[1].set_title("MSE over calls")
            fig.tight_layout(); plt.show()

    def _on_change(change):
        if change["name"] == "value":
            render(change["new"])

    dd.observe(_on_change)
    display(dd)
    render(dd.value)
    display(out)
```
